```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  # uppercase to lowercase
  text = str(text).lower()
  # remove the punctuation
  text = re.sub(r'[^a-zA-Z\s]','',text)
  # first perform tokenization then - remove stopWords
  words = text.split()
  words = [word for word in words if word not in stop_words]
  # lemmatize
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
df_ML['tags'] = df_ML['tags'].apply(preprocess_text)
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
indices
```

```python
# Apply the Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000, ngram_range=(1,2) , stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df_ML['tags'])
```

```python
tfidf_matrix
```



Now we complete the matrix , tfidf , indecies
Now we will create are Machine Learning Model using Cosine similarity




```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10):
  if title not in  indices:
    return ['Movie not found']

  idx = indices[title]

  sim_score = cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('The Dark Knight')
```
